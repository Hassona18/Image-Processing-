Here's a breakdown of the *algorithms names* for each process described in the project:

---

### *Image Color Conversion*
1. *Grayscale Conversion*  
   Algorithm: Weighted sum method using luminance formula \((0.299R + 0.587G + 0.114B)\).

---

### *Thresholding*
1. *Threshold Calculation*  
   Algorithm: Global Thresholding using the average value of pixel intensities.

2. *Optimal Thresholding Evaluation*  
   Algorithm: Otsu's Method (for evaluating optimal threshold).

---

### *Halftone*
1. *Simple Halftone*  
   Algorithm: Binary thresholding.

2. *Advanced Halftone*  
   Algorithm: Floyd-Steinberg Error Diffusion.

---

### *Histogram Operations*
1. *Histogram Calculation*  
   Algorithm: Pixel intensity frequency count.

2. *Histogram Equalization*  
   Algorithm: Cumulative Distribution Function (CDF) for intensity remapping.

---

### *Simple Edge Detection Methods*
1. *Sobel Operator*  
   Algorithm: Sobel kernels for horizontal and vertical gradients.

2. *Prewitt Operator*  
   Algorithm: Prewitt kernels for horizontal and vertical gradients.

3. *Kirsch Compass Masks*  
   Algorithm: Kirsch mask-based gradient calculation in eight compass directions.

---

### *Advanced Edge Detection Methods*
1. *Homogeneity Operator*  
   Algorithm: Local contrast-based edge detection (difference between max and min pixel intensities in a region).

2. *Difference Operator*  
   Algorithm: Gradient calculation using directional difference masks.

3. *Difference of Gaussians (DoG)*  
   Algorithm: Subtraction of two blurred images using 7×7 and 9×9 Gaussian masks.

4. *Contrast-based Edge Detection*  
   Algorithm: Smoothing mask from figure 6.10.

5. *Variance Edge Detection*  
   Algorithm: Local variance in pixel intensities.

6. *Range Edge Detection*  
   Algorithm: Difference between maximum and minimum pixel values in a region.

---

### *Filtering*
1. *High-pass Filter*  
   Algorithm: High-frequency enhancement mask from figure 7.18.

2. *Low-pass Filter*  
   Algorithm: Averaging filter mask from figure 7.3.

3. *Median Filter*  
   Algorithm: Replace each pixel with the median value of its neighborhood.

---

### *Image Operations*
1. *Copy and Add Images*  
   Algorithm: Pixel-wise addition.

2. *Copy and Subtract Images*  
   Algorithm: Pixel-wise subtraction.

3. *Invert Image*  
   Algorithm: Inversion formula \(255 - \text{pixel value}\).

---

### *Histogram-Based Segmentation*
1. *Manual Technique*  
   Algorithm: Manually select threshold levels.

2. *Histogram Peak Technique*  
   Algorithm: Use peaks in histogram for segmentation.

3. *Histogram Valley Technique*  
   Algorithm: Use valleys in histogram for segmentation.

4. *Adaptive Histogram Technique*  
   Algorithm: Local thresholding based on intensity variations in regions.

---
